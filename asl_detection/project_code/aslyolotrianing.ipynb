```python
import os
import matplotlib.pyplot as plt
from collections import Counter

# Define the paths for train and val datasets
train_path = "yolo_dataset/images/train"  # Update with your actual path
val_path = "yolo_dataset/images/val"  # Update with your actual path

# Get the list of image files in the train and validation directories
train_images = [f for f in os.listdir(train_path) if os.path.isfile(os.path.join(train_path, f))]
val_images = [f for f in os.listdir(val_path) if os.path.isfile(os.path.join(val_path, f))]

# Count the number of images in each dataset
num_train_images = len(train_images)
num_val_images = len(val_images)

# Assuming each image has a corresponding label in 'yolo_dataset/labels/train' and 'yolo_dataset/labels/val'
train_labels_path = "yolo_dataset/labels/train"  # Update with your actual path
val_labels_path = "yolo_dataset/labels/val"  # Update with your actual path

# Count the number of occurrences of each class in the train and validation labels
train_labels = [f.split('.')[0] for f in os.listdir(train_labels_path)]
val_labels = [f.split('.')[0] for f in os.listdir(val_labels_path)]

# Count the occurrences of each class
train_class_counts = Counter(train_labels)
val_class_counts = Counter(val_labels)

# Calculate the class distribution
train_class_distribution = dict(train_class_counts)
val_class_distribution = dict(val_class_counts)

# Plot the bar charts for class distribution in training and validation datasets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot for training data
ax1.bar(train_class_distribution.keys(), train_class_distribution.values(), color='blue')
ax1.set_title("Training Dataset Class Distribution")
ax1.set_xlabel("Class")
ax1.set_ylabel("Number of Images")
ax1.set_xticklabels(train_class_distribution.keys(), rotation=45)

# Plot for validation data
ax2.bar(val_class_distribution.keys(), val_class_distribution.values(), color='orange')
ax2.set_title("Validation Dataset Class Distribution")
ax2.set_xlabel("Class")
ax2.set_ylabel("Number of Images")
ax2.set_xticklabels(val_class_distribution.keys(), rotation=45)

# Display the plots
plt.tight_layout()
plt.show()

num_train_images, num_val_images, train_class_distribution, val_class_distribution
```

```python
import os
from collections import Counter

# Define the paths for train and val datasets
train_path = "yolo_dataset/images/train"  # Update with your actual path
val_path = "yolo_dataset/images/val"  # Update with your actual path

# Get the list of image files in the train and validation directories
train_images = [f for f in os.listdir(train_path) if os.path.isfile(os.path.join(train_path, f))]
val_images = [f for f in os.listdir(val_path) if os.path.isfile(os.path.join(val_path, f))]

# Extract class names based on the first letter of the filename
train_classes = [f[0] for f in train_images]  # Take the first letter of the filename
val_classes = [f[0] for f in val_images]  # Take the first letter of the filename

# Combine both train and validation classes
combined_classes = train_classes + val_classes

# Count the occurrences of each class in the combined dataset
class_counts = Counter(combined_classes)

# Calculate the total number of images in train, validation, and combined datasets
num_train_images = len(train_images)
num_val_images = len(val_images)
num_total_images = num_train_images + num_val_images

# Print the class distribution and total image counts
print(f"Total number of images: {num_total_images}")
print(f"Total number of images in the training dataset: {num_train_images}")
print(f"Total number of images in the validation dataset: {num_val_images}")
print("\nClass Distribution (Total Images per Class):")
for class_name, count in class_counts.items():
    print(f"Class {class_name}: {count} images")
```
